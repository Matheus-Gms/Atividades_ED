### Atividade 01 - Lista Sequêncial

```python
#Criação da classe lista
class ListaSeq:
    def __init__(self, tamMax):
        self.elementos = []
        self.tamMax = tamMax
        self.tamAtual = 0

    def tamanhoMax(self):
        return self.tamMax
    
    #Verificar se a lista está vazia
    def ListaVazia(self):
        if not self.elementos:
            print ("A lista está vazia")
        else:
            print("A lista contém os seguintes elementos: ", self.elementos)
    
    #Verificar se a lista está cheia
    def ListaCheia(self):
        if (self.tamAtual == self.tamMax):
            print ("A lista está cheia")
        else:
            print("A lista ainda contém espaços vazios")
    
    #Verificar tamanho da lista
    def tamLista(self):
        return self.tamAtual
         
    #Visuzalizar a lista atual
    def printLista(self):
        return self.elementos
    
    #Verificar um elemento em uma posição específica da lista
    def verificarElemento(self, pos):
        if (pos > self.tamLista()+1 or pos < 1):
            print (f"\nA psição {pos} é inválida")
        else:
            print(f"\nElemento na posição {pos} da lista: {self.elementos[pos-1]}") 
            
    #Modificar um elemento em uma posição específica da lista
    def modificarElemento(self, pos, elemento):
        if (pos > self.tamLista()+1 or pos < 1):
            print(f"\nA posição {pos} é inválida")
        else:   
            print(f"Modificando o elemento da posição {self.elementos[pos]} pelo valor {elemento}")
            self.elementos[pos-1] = elemento
        return self.elementos

    #Adicionar um elemento em uma posição específica da lista
    def addPos(self, pos, elemento):
        if (self.tamLista() == self.tamanhoMax()):
            print ("A lista está cheia\n")
            
        elif (pos > self.tamAtual+1 or pos <= 0):
            print(f"\nAdicionando na posição {pos} o elemento {elemento}")
            print (f"A posição {pos} é inválida")
            
        else:
            self.elementos.append(0)
            for i in range (len(self.elementos)-1, pos-1, -1):     
                self.elementos[i] = self.elementos[i-1] 
            self.elementos[pos-1] = elemento
            self.tamAtual = self.tamAtual + 1
            print(f"\nAdicionando na posição {pos} o elemento {elemento}")
            
   #Remover um elemento de uma posição específica da lista
    def removerElemento(self, pos):
        if (self.tamLista() == 0):
            print ("A Lista está vazia")
        elif (pos > self.tamAtual+1 or pos < 1):
            print (f"\nA posição {pos} é inválida")
        else:
            for i in range(pos-1, self.tamAtual-1, +1):
                self.elementos[i] = self.elementos[i+1] 
            self.elementos = self.elementos[:-1]
            self.tamAtual = self.tamAtual - 1
            print(f"\nRemovendo o elemento da posição {pos}")
```
