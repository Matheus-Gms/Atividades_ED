### Atividade 02 - Lista Encadeada

```python
class No:
    def __init__(self, dado):
        self.dado = dado
        self.prox = None
```

```python
class Lista:
    def __init__(self):
        self.cabeca = None
        self.tamanho = 0
        
    def tamanho_(self):
        return self.tamanho
    
    def vazia(self):
        if not self.tamanho:
            return True
        return False
    
    def insere(self, pos, elemento):
        cont = 1
        if (pos == 1):
            self.cabeca = No(elemento)
            self.tamanho += 1
            
        elif ((pos > self.tamanho) or (pos < 1)):
            print("Posição inválida")
        
        else:
            aux = self.cabeca
            while (cont < pos-1) or (aux != None):
                aux = aux.prox
                cont += 1
            aux.prox = No(elemento)
            self.tamanho += 1
            return True
    
    def busca(self, pos):
        aux = self.cabeca
        #cont = 1
        
        if((pos < 1) or (pos > self.tamanho)):
            print("Posição inválida.")
        
        elif (pos - 1 == 0):
            return aux

        else:
            for i in range(pos):
                aux = aux.prox
                cont += 1
            return aux.dado
        
    def troca(self, pos, elemento):
        aux = self.cabeca
        cont = 0
        
        if((pos - 1 < 0) or (pos > self.tamanho)):
            print("Posição inválida.")
        
        else:
            if (pos - 1 == 0):
                self.cabeca = elemento
            else:
                while(cont != pos - 1):
                    aux = aux.prox
                    cont += 1
                aux.dado = elemento
                
    def remove(self, pos):
        aux = self.cabeca
        cont = 0
        
        if (self.vazia == True):
            print("Lista vazia")
        
        elif (pos - 1 == 0):
            aux = None
            self.tamanho -= 1
            
        elif (pos - 1 < 0 or pos > self.tamanho):
            print("Posição inválida")
        
        else:
            while(cont != pos - 1):
                aux = aux.prox
                cont += 1
            aux.dado = aux.prox
            self.tamanho -= 1  

    def __str__(self):
        listaAux = []
        add = None
        aux = self.cabeca
        if(aux == None):
            return False
        else:
            while (aux != None):
                add = aux.dado
                listaAux.append(add)
                aux = aux.prox
            return listaAux
```
